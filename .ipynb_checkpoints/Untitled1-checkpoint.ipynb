```python
import pandas as pd
excel_File = 'News Stories.xlsx'

book = pd.read_excel(excel_File, index_col=None,usecols=['Headline','Category','Story'])

pakistan=0
sport=0
science=0
lollywood=0
near_Ones=0
world=0
longest_Story=''
smallest_Story=999999
collec = [];


for row in book.iterrows():
    
    #getting elements from excel file and removing extra whitespaces
    title=row[1][0]
    title=title.strip()
    story=row[1][2]
    story=story.strip()
    category=row[1][1]
    category=category.strip()
    
    
    
    
    #replace all special characters
    single_Entity = title+" "+" "+story+" "+category
    single_Entity =single_Entity.replace(':','')
    single_Entity =single_Entity.replace('‘','')
    single_Entity =single_Entity.replace('؟','')
    single_Entity =single_Entity.replace('،','')
    single_Entity =single_Entity.replace('’','')
    single_Entity =single_Entity.replace('۔','')
    
    words=[]
    words=single_Entity.split(' ')
    
    
    for word in words:
        if not collec.count(word):
            collec.append(word)
            
    
    if len(longest_Story)<len(story):
        longest_Story = story
    if 100000 > len(story):
        smallest_Story=''
        smallest_Story=story
    if  category=='پاکستان':
        pakistan+=1
    elif  category=='آس پاس':
        near_Ones+=1
    elif  category=='ورلڈ':
        world+=1
    elif  category=='کھیل':
        sport+=1
    elif  category=='فن فنکار':
        lollywood+=1
    elif  category=='سائنس':
        science+=1
        
print(len(collec))
          
        
   

```
